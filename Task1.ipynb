```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Iris.csv'
iris_data = pd.read_csv(file_path)
```

```python
# Aggregate the data to get mean values for each species
species_mean = iris_data.groupby('Species').mean()
```

```python
# Define colors for each feature
colors = ['orange', 'red', 'pink', 'blue']

# Plot the bar chart for each feature
ax = species_mean.plot(kind='bar', figsize=(12, 8), color=colors)
plt.title('Mean Values of Iris Features by Species')
plt.xlabel('Species')
plt.ylabel('Mean Values')
plt.xticks(rotation=0)
plt.legend(title='Features')
plt.show()
```

```python
# Define the features
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Create scatter plots with regression lines for each pair of features using different colors
fig, axes = plt.subplots(3, 2, figsize=(15, 20))
axes = axes.flatten()

# Plot scatter plots with regression lines
plot_idx = 0
colors = ['red', 'blue', 'green', 'purple', 'pink', 'brown']  # Different colors for each feature pair
for i, feature1 in enumerate(features):
    for j, feature2 in enumerate(features):
        if feature1 != feature2 and plot_idx < len(axes):
            ax = axes[plot_idx]
            color = colors[plot_idx % len(colors)]
            sns.regplot(data=iris_data, x=feature1, y=feature2, ax=ax, scatter_kws={'s': 20}, line_kws={'color': color}, fit_reg=True)
            sns.scatterplot(data=iris_data, x=feature1, y=feature2, hue='Species', ax=ax, palette='deep', legend=False)
            plot_idx += 1

plt.suptitle('Scatter Plots with Regression Lines of Iris Features')
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

```python
# Define the features
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Create box plots for each feature
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

# Plot the box plots
for i, feature in enumerate(features):
    ax = axes[i]
    sns.boxplot(data=iris_data, x='Species', y=feature, hue='Species', ax=ax, palette=['orange', 'red', 'pink'])
    ax.set_title(f'Box plot of {feature} by Species')
    if ax.get_legend() is not None:
        ax.get_legend().remove()  # Remove the legend as it's redundant

plt.suptitle('Box Plots of Iris Features by Species')
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```
